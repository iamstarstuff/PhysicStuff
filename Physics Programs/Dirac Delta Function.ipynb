## Dirac Delta using Gaussian function
In statistics we often use so-called Gaussian curves, of the form 
$$f(x)=\dfrac{e^{-\frac{x^2}{2\sigma^2}}}{\sigma \sqrt{2\pi}}$$
The width of the Gaussian is controlled by $\sigma$. For $f$ as a probability density, $f(x)dx$ would be probability of $f$ lying between $x$ and $x+dx$. Since $\int_{-\infty}^\infty e^{-\lambda x^2}dx=\sqrt{\pi/\lambda}$, there is unit area under the function given by the above expression, indicating that the probability of $f$ lying between $-\infty$ and $\infty$ is $1$.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [8,5]
plt.rcParams['figure.dpi'] = 600

def Delta(sigma,x):
    return (np.exp(-x**2/(2*sigma**2))/(sigma*np.sqrt(2*np.pi)))

x = np.arange(-10,10,0.01)
sigma = [0.3,0.4,0.5,1,2,3,4]
for i in sigma:
    y = []
    for j in x:
        y.append(Delta(i,j))
    plt.plot(x,y,label=f'$\sigma$={i}')
plt.grid()
plt.legend()
    
    
```

## Using `SciPy`
`scipy.signal.unit_impulse()` Unit impulse signal(discrete delta function) or unit basis vector.

```python
import numpy as np 
from scipy import signal
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [8,5]
plt.rcParams['figure.dpi'] = 600
impulse = signal.unit_impulse(200,100)
plt.plot(np.arange(-100,100),impulse)
plt.xlabel('Time[samples]')
plt.ylabel('Amplitude')
plt.grid()
```
