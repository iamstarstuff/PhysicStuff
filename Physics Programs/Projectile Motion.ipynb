# Path of Projectile

## Important equations for projectile motion: 

## $\text{Horizontal components}$  
### $v_x=u\cdot\cos(\theta)$  
### $x = u\cdot\cos(\theta)\cdot t$
### $v_x^2=u^2\cdot\cos^2(\theta)$  


## $\text{Vertical components}$  
### $v_y=u\cdot\sin(\theta)-gt$  
### $y=u\cdot\sin(\theta)\cdot t - \dfrac{1}{2}gt^2$  
### $v_y^2=u^2\cdot\sin^2(\theta)-2gy$  


### $\text{Time of flight:}\;\;\;\;T=\dfrac{2u\cdot\sin(\theta)}{g}$


### $\text{Range:}\;\;\;\;R=\dfrac{u^2\sin(2\theta)}{g}$


### $\text{Maximum Height:}\;\;\;\;H=\dfrac{u^2\sin^2(\theta)}{2g}$

```python
import numpy as np

# Acceleration due to gravity in m/s
g = 9.8  

def x_component(u,theta,time):
    """
    Arguments:
        u : initial velocity in m/s
        theta : angle from horizontal in degrees 
    Returns:
        Horizontal component of motion
    """
    #max_time = (2*u*np.sin(np.deg2rad(theta)))/g #max time
    #if time <= max_time:
    return u*np.cos(np.deg2rad(theta))*time
    #else:
    #    print("The time is more than time of flight of projectile")

def y_component(u,theta,time):
    """
    Arguments:
        u : initial velocity in m/s
        theta : angle from horizontal in degrees
    Returns:
        Vertical component of motion
    """
    #max_time = (2*u*np.sin(np.deg2rad(theta)))/g #max time
    #if time <= max_time:
    return u*np.sin(np.deg2rad(theta))*time - 0.5*g*time**2
    #else:
    #    print("The time is more than time of flight of projectile")

def time_of_flight(u,theta):
    """
    Arguments:
        u : initial velocity in m/s
        theta : angle from horizontal in degrees
    Returns :
        Time of flight of projectile
    """
    return (2*u*np.sin(np.deg2rad(theta)))/g

def range_of_projectile(u,theta):
    """
    Arguments :
        u : initial velocty in m/s
        theta : angle from horizontal in degrees
    Returns:
        Range of projectile
    """
    return (u**2*np.sin(2*np.deg2rad(theta)))/g

def max_height(u,theta):
    """
    Arguments :
        u : initial velocity in m/s
        theta : angle from horizontal in degrees
    Returns :
        Maximum height of projectile
    """
    return (u**2*(np.sin(np.deg2rad(theta)))**2)/(2*g)

def area_under_path(u,theta):
    """
    Arguments :
        u : initial velocty in m/s
        theta : angle from horizontal in degrees
    Returns:
        Area under the path of projectile
    """
    t = np.deg2rad(theta)
    return (2*u**4*(np.sin(t))**3*np.cos(t))/(3*g**2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook
import matplotlib.image as image
plt.rcParams['figure.figsize']=[15,10]
plt.rcParams['figure.dpi']=300



time1 = time_of_flight(60, 30)*np.linspace(0,1,100)
x1 = x_component(60,30,time1)
y1 = y_component(60,30,time1)


time2 = time_of_flight(30, 30)*np.linspace(0,1,100)
x2 = x_component(30,30,time2)
y2 = y_component(30,30,time2)

fig,ax=plt.subplots()
ax.plot(x1,y1,label="For $u=60 m/s$")
ax.plot(x2,y2,label="For $u=30 m/s$")
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.legend()
ax.grid()
fig.text(0.5,0.5,'Â©PhysicStuff 2021',ha='center',va='center',color='gray',fontsize=80,alpha=0.10)
```

```python
u = [20,30,40,50,60,70]
time,x,y = [[] for i in range(len(u))],[[] for i in range(len(u))],[[] for i in range(len(u))]

fig,ax = plt.subplots()

for i in range(len(u)):
    time[i] = time_of_flight(u[i],40)*np.linspace(0,1,100)
    x[i] = x_component(u[i],40,time[i])
    y[i] = y_component(u[i],40,time[i])
    ax.plot(x[i],y[i],label=f"For $u={u[i]}\;m/s$")
    ax.annotate(f'For {u[i]}m/s\n{round(x[i][-1],1)} m',xy=(x[i][-1],0),xytext=(x[i][-1]+5,y[i][-1]+5),arrowprops=dict(arrowstyle="->",linewidth=0.5))
ax.legend()
ax.grid(axis='both')
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
fig.text(0.5,0.5,'Â©PhysicStuff 2021',ha='center',va='center',color='gray',fontsize=80,alpha=0.10)
plt.title("Path of projectile with $\\theta=40^\circ$ and varying initial velocity")
```
