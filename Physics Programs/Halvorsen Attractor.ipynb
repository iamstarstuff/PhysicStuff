# Halvorsen Attractor
## $$\dfrac{dx}{dt}=-ax-4y-4z-y^2$$
## $$\dfrac{dy}{dt}=-ay-4z-4x-z^2$$
## $$\dfrac{dz}{dt}=-az-4x-4y-x^2$$
### The values of constants are   
### $$a = 1.4$$

```python
import matplotlib.pyplot as plt
import numpy as np
# Use `%matplotlib notebook` so that you can view/scroll in 3D within the notebook
%matplotlib inline

def halvorsen(x,y,z, a=1.89):
    x_dot = -a*x-4*y-4*z-y**2
    y_dot = -a*y-4*z-4*x-z**2
    z_dot = -a*z-4*x-4*y-x**2
    return x_dot,y_dot,z_dot

dt = 0.01
num_steps = 100000

x = np.empty(num_steps+1)
y = np.empty(num_steps+1)
z = np.empty(num_steps+1)

# Set initial values
x[0], y[0], z[0] = (-1.48, -1.51, 2.04)

x_values = []
y_values = []
z_values = []

# update values of x,y,z
for i in range(num_steps):
    x_dot, y_dot, z_dot = halvorsen(x[i],y[i],z[i])
    x[i + 1] = x[i] + (x_dot * dt)
    y[i + 1] = y[i] + (y_dot * dt)
    z[i + 1] = z[i] + (z_dot * dt)
    
    x_values.append(x[i])
    y_values.append(y[i])
    z_values.append(z[i])
    

ax = plt.figure(figsize=(8,8),dpi=300).add_subplot(projection='3d')
ax.plot(x_values, y_values, z_values, lw=0.03)
#ax.set_xlabel("X Axis")
#ax.set_ylabel("Y Axis")
#ax.set_zlabel("Z Axis")
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.set_axis_off()
ax.set_title("Halvorsen Attractor",fontsize=10)
ax.view_init(-34,-136)
plt.show()
```
