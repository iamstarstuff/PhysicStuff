# Henon Map
Also called as **Henon-Pomeau attractor/map** is a discrete-time dynamical system. The Henon map takes a point $(x_n,y_n)$ in the plane and maps it to new point
$\begin{cases}
    x_{n+1} = 1-ax_n^2+y_n\\
    y_{n+1}=bx_n
\end{cases}$  

The map depends on two parameters, $a$ and $b$ which for the **Classical Henon Map** have values of $a=1.4$ and $b=0.3$. For the classical values the Henon Map is chaotic.

```python
import numpy as np
import matplotlib.pyplot as plt
import mplcyberpunk
%matplotlib inline

def henon(x,y,a,b):
    """
    x and y are initial values
    a and b are parameters which are 1.4 and 0.3 resp for classical henon map
    Call this function in a loop to get points for Henon Map
    """
    return 1-a*x**2+y,b*x
```

```python
x = 0
y = 0
x_values = []
y_values = []

for i in range(1000000):
    x,y = henon(x,y,1.4,0.3)
    x_values.append(x)
    y_values.append(y)

plt.figure(figsize=(15,15),dpi=600)
plt.style.use("cyberpunk")
plt.scatter(x_values,y_values,c='y',s=0.1,lw=0.01)
plt.title("Henon Map",fontsize=30)
```
