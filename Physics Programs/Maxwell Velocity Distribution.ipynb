# Maxwell Velocity Distribution
Number of molecules in the gas at thermal equilibrium at temperature T, having speed u in a unit solid angle is given by:  
$$f_v=\Big(\dfrac{2}{\pi}\Big)^{\frac{1}{2}}\Big(\dfrac{m}{kT}\Big)^{\frac{3}{2}}u^2\,exp\Big(-\dfrac{mu^2}{2kT}\Big)$$  
Take range of u with 100 m/s to 1000 m/s  
Plot the function for T = 250 K, 300 K, 500 K, for $m$ of nitrogen molecule  
Plot the function for T = 300 K and with $m$ for hydrogen and carbon dioxide molecules

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import pi,k

def fv(u,m,T):
    return ((2/pi)**0.5)*((m/k/T)**1.5)*(u**2)*(np.exp(-m*u**2/2/k/T))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import pi,k

u = np.linspace(100,1000,100)
T = [250,300,500]
mass = [4.65283e-26,3.347646e-27,7.308032e-26] # masses of N2,H2 and CO2 resp in kilograms

def fv(u,m,T):
    return ((2/pi)**0.5)*((m/k/T)**1.5)*(u**2)*(np.exp(-m*u**2/2/k/T))
```

```python
# for Nitrogen
fN = [[] for i in range(len(T))] # creating empty list for each T
for i in range(len(T)):
    for j in range(len(u)):
        fN[i].append((fv(u[j],mass[0],T[i])))
plt.figure(figsize=(10,5),dpi=600)
plt.plot(u,fN[0],label="250 K")
plt.plot(u,fN[1],label="300 K")
plt.plot(u,fN[2],label="500 K")
plt.legend()
plt.title("For Nitrogen Molecule")
plt.xlabel("Velocity in $m/s$")
plt.ylabel("Ratio of molecules having that velocity") 
```

```python
# For Hydrogen Molecule
fH = [[] for i in range(len(T))]
for i in range(len(T)):
    for j in range(len(u)):
        fH[i].append((fv(u[j],mass[1],T[i])))
plt.figure(figsize=(10,5),dpi=600)
plt.plot(u,fH[0],label="250 K")
plt.plot(u,fH[1],label="300 K")
plt.plot(u,fH[2],label="500 K")
plt.legend()
plt.title("For Hydrogen Molecule Molecule")
plt.xlabel("Velocity in $m/s$")
plt.ylabel("Ratio of molecules having that velocity")
```

```python
# For Carbon Dioxide Molecule
fCD = [[] for i in range(len(T))]
for i in range(len(T)):
    for j in range(len(u)):
        fCD[i].append((fv(u[j],mass[2],T[i])))
plt.figure(figsize=(10,5),dpi=600)
plt.plot(u,fCD[0],label="250 K")
plt.plot(u,fCD[1],label="300 K")
plt.plot(u,fCD[2],label="500 K")
plt.legend()
plt.title("For Carbon Dioxide Molecule")
plt.xlabel("Velocity in $m/s$")
plt.ylabel("Ratio of molecules having that velocity")
```
