# Fourier Synthesis of waves
Any steady state wave is composed of a series of sinusoidal components whose frequencies are those of the fundamental and its harmonics, each component having proper amplitude and phase. One way of giving phase information for a particular frequency is giving amplitudes of sine and cosine components of the same frequency.  

In **Fourier Synthesis**, the function is generated by adding sine and cosine components.  
Consider the following functions:  
$$f_1(x)=\sum_{n}\dfrac{\sin nx}{n}\quad \quad n = 1,3,5,...\quad \quad \text{Add 10 terms}$$
$$f_2(x)=\sum_{n}\dfrac{\cos nx}{n^2}\quad \quad n = 1,3,5,...\quad \quad \text{Add 10 terms}$$
$$f_3(x)=1-2\sum_{n}\dfrac{\cos 2nx}{(4n^2-1)}\quad \quad n = 1,2,3,...\quad \quad\text{Add 10 terms}$$
Plot $f_1(x),\,f_2(x),\,f_3(x)\;\text{ for }\;x\in[0,6\pi]$

```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    sum = 0
    for n in range(20):
        if n%2!=0:
            sum+= np.sin(n*x)/n
        else:
            continue
    return sum


def f2(x):
    sum = 0
    for n in range(20):
        if n%2!=0:
            sum+= np.cos(n*x)/n**2
        else:
            continue
    return sum

def f3(x):
    sum = 0
    for n in range(1,11):
        sum+= np.cos(2*n*x)/(4*n**2-1)
    return 1-2*sum

x = np.linspace(0,6*np.pi,10000)
fig, (ax1,ax2,ax3) = plt.subplots(3,1,figsize=(15,10),sharex=True)

ax1.plot(x,f1(x),c='r',label='$f_1(x)=\sum_{n}\dfrac{\sin nx}{n}$')
ax1.legend()
ax2.plot(x,f2(x),c='g',label='$f_2(x)=\sum_{n}\dfrac{\cos nx}{n^2}$')
ax2.legend()
ax3.plot(x,f3(x),c='b',label='$f_3(x)=1-2\sum_{n}\dfrac{\cos 2nx}{(4n^2-1)}$')
plt.xlabel('X')
ax3.legend()
plt.show()
```
