# Area of Shaded Region - Infinite Squares & Circles
This was taken from an instagram page [@mymathssolutions](https://www.instagram.com/mymathssolutions/)

```python
import numpy as np

def areaofshaded(side:'length of first square',n:'no. of iterations',units:'cm,m,ft,units,whatever'):
    s = []  # list to store side lengths
    s.append(side)   # append first length
    a_s = []   # list for areas of square
    a_c = []   # list for areas of circles
    shaded = 0   # will add area of shaded region to this variable
    for i in range(0,n+1):
        
        # The diameter of first circle is equal to diagonal of second square,  
        # so side of consecutive square will be 1/sqrt(2) times
        
        s.append(s[i]/np.sqrt(2))  # making list of side lengths of square
        
        a_s.append(s[i]**2)  # appending areas of sqare
        
        a_c.append(np.pi*(s[i]/2)**2)   # appending areas of circles
        
        shaded += a_s[i]-a_c[i]   # adding the shaded areas
        
    print(f"Side of first square is {side} {units}, for {n} iterations, the\
 blue area (area of shaded region) is \n{shaded} sq.{units}.") 
```

```python
areaofshaded(2,10,'cm')
```

```python
areaofshaded(2,100,'units')
```

```python
areaofshaded(2,1000,'cm')
```
