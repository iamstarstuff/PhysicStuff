# Dini's surface

A surface of constant negative curvature obtained by twisting a pseudosphere is known as Dini's Surface. It is named after an Italian mathematician _Ulisse Dini_  
Another description is a helicoid constructed from the tractrix.
<center>The parametric equations are</center>

## $$x = a\,cos(u)sin(v)$$
## $$y = a\,sin(u)sin(v)$$
## $$z = a(cos(u)+ln(tan\frac{v}{2})+bu$$
### $$where\;,\;\;0\leq u \leq 4\pi\;\;\;\;0<v<1$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['figure.dpi']=300
%matplotlib inline
```

```python
def dini(a,b):
    u = np.linspace(0,4*np.pi,2000)
    v = np.linspace(0.01,1,2000)
    u,v = np.meshgrid(u,v)
    x = a*np.cos(u)*np.sin(v)
    y = a*np.sin(u)*np.sin(v)
    z = a*(np.cos(v)+np.log(np.tan(v/2)))+(b*u)
    return x,y,z
```

```python
x,y,z=dini(1,0.5)
```

```python
fig,ax = plt.subplots(1,2,dpi=1200,subplot_kw={"projection":"3d"},constrained_layout=True)
ax[0].plot_surface(x,y,z,linewidth=0.15,cmap='viridis')
ax[1].plot_surface(x,y,z,linewidth=0.15,cmap='viridis')
ax[0].set_axis_off()
ax[1].set_axis_off()
ax[0].view_init(30,10)
ax[1].view_init(10,0)
```

```python
fig = plt.figure(dpi=600)
ax = plt.axes(projection='3d')
plt.style.use('dark_background')
ax.plot_surface(x,y,z,linewidth=0.15,cmap='viridis')
ax.set_axis_off()
ax.view_init(30,10)
```
