# Collatz Conjecture
Consider following operation on an arbitrary positive integer:
- If number is even divide it by 2
- If number is odd triple it and add 1

$$
\begin{equation}
f(n) =
    \begin{cases}
        \dfrac{n}{2}  & \quad \text{if } n \equiv 0 \\
        3n+1    & \quad \text{if } n \equiv 1 
    \end{cases}
\end{equation}
$$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(n):
    x_list = []
    count_list = []
    count = 0
    while n!=1:
        x_list.append(n)
        count_list.append(count)
        count+=1
        if n%2==0:
            n = n/2
        elif n%2!=0:
            n = 3*n+1
        
    print(f"For n = {x_list[0]}\nIt took {count} steps to reach 4-2-1 loop.")
    print(f"The maximum value the sequence attains is {int(max(x_list))}")
    plt.figure(figsize=(10,5),dpi=600)
    plt.plot(count_list,x_list,'--',lw=0.5,marker='o',ms=2,label=f"For n = {x_list[0]}")
    plt.grid(True)
    plt.legend()
    plt.xlabel("No. of steps to reach 4-2-1 loop")
    plt.ylabel(f"Values of sequence for n = {x_list[0]}")
    
    
```

```python
# To plot sequence of a list of numbers in the same plot
def fn(n):
    master_x_list = []
    master_count_list = []
    
    for i in range(len(n)):
        nt = n[i]
        
        x_list = []
        count_list = []
        count = 0
        
        while nt!=1:
            x_list.append(nt)
            count_list.append(count)
            count+=1
            if nt%2==0:
                nt = nt/2
            elif nt%2!=0:
                nt = 3*nt+1
                
        print(f"For n = {x_list[0]}\nIt took {count} steps to reach 4-2-1 loop.")
        print(f"The maximum value the sequence attains is {int(max(x_list))}\n\n")
        master_x_list.append(x_list)
        master_count_list.append(count_list)
        
    plt.figure(figsize=(10,5),dpi=600)
    for i in range(len(master_x_list)):
        plt.plot(master_count_list[i], master_x_list[i], label=f"For n = {master_x_list[i][0]}")
    plt.legend()
    plt.grid()
                
    
```

# Checking which n attains maximum value 

```python
def maxfn(n):
    master_x_list = []
    master_count_list = []
    maximum = []
    
    for i in range(len(n)):
        nt = n[i]
        
        x_list = []
        count_list = []
        count = 0
        
        while nt!=1:
            x_list.append(nt)
            count_list.append(count)
            count+=1
            if nt%2==0:
                nt = nt/2
            elif nt%2!=0:
                nt = 3*nt+1
                
        #print(f"For n = {x_list[0]}\nIt took {count} steps to reach 4-2-1 loop.")
        #print(f"The maximum value the sequence attains is {int(max(x_list))}\n\n")
        master_x_list.append(x_list)
        master_count_list.append(count_list)
        maximum.append(int(max(x_list)))
        
    #print(f"Maximums for given list: {maximum}")
    index = np.argmax(maximum)
    print(f"For given list of values we get the highest maximum of {int(max(maximum))} for n = {int(n[index])}")        
    
```

```python
maxfn([73,97,129,171,231])
```

```python
maxfn([4,5,6,7,8,100,81])
```

# Geometric Brownian Motion
Aka exponential Brownian motion, is a continuous-time stochastic process in which the logarithm of the randomly varying quantity follows a Brownian motion with drift. 

```python
# Example code from wikipedia
import numpy as np 
import matplotlib.pyplot as plt 
mu = 1
n = 50
dt = 0.1
x0 = 100

sigma = np.arange(0.8,2,0.2)

x = np.exp((mu - sigma**2/2)*dt + sigma*np.random.normal(0,np.sqrt(dt),size=(len(sigma),n)).T)
x = np.vstack([np.ones(len(sigma)),x])
x = x0*x.cumprod(axis=0)

plt.figure(figsize=(10,5),dpi=300)
plt.plot(x)
plt.legend(np.round(sigma,2))
plt.xlabel("$t$")
plt.ylabel("$x$")
plt.title("Realisation of Geometric Brownian Motion with different variances\n$\mu=1$")
```
